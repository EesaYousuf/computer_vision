```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = sns.load_dataset('tips')
print(df.head())
```

1. Violin Plot
Violin plots show the distribution of data across categories and combine boxplot + KDE


```python
sns.violinplot(x='day', y='total_bill', data=df, palette='Set2')
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

2.Swarm Plot
Swarm plots show all data points and avoid overlapping using a jittering algorithm.



```python
sns.swarmplot(x='day', y='total_bill', data=df, color='black')
plt.title('Swarm Plot of Total Bill by Day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=df, inner=None, palette='Set2')
sns.swarmplot(x='day', y='total_bill', data=df, color='k', alpha=0.6)
plt.title('Violin + Swarm Plot of Total Bill by Day')
plt.show()
```
