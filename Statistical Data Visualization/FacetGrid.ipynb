```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load sample dataset
tips = sns.load_dataset("tips")
```

1. Using FacetGrid
FacetGrid creates a matrix of subplots based on the values of one or more categorical variables.


```python
# Create a FacetGrid by 'sex' and 'time'
g = sns.FacetGrid(tips, col="sex", row="time")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
plt.show()
```

 2. Count Plot with FacetGrid

```python
g = sns.FacetGrid(tips, col="sex")
g.map_dataframe(sns.countplot, x="day")
plt.show()
```

```python
# 3. Box Plot with FacetGrid

g = sns.FacetGrid(tips, col="day", height=4, aspect=0.9)
g.map_dataframe(sns.boxplot, x="sex", y="total_bill")
plt.show()
```

```python
#  4. Bar Plot with FacetGrid

g = sns.FacetGrid(tips, col="time")
g.map_dataframe(sns.barplot, x="sex", y="tip", ci=None)
plt.show()
```
